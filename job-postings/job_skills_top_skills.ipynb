```python
import pandas as pd
import numpy as np
```

```python
skills = pd.read_csv("job_skills.csv")
```

```python
#seperate the skills and put them into lists
skills["job_skills"] = skills["job_skills"].str.split(",")
skills.head()
```

```python
job_skills = list(skills["job_skills"])
type(job_skills[0][0])
```

```python
from collections import Counter
```

```python
job_skills
```

```python
word_counts = Counter()
```

```python
for element in job_skills:
    # Check if the element is a list
    if isinstance(element, list):
        # Flatten the list and count occurrences of each word
        word_counts.update(element)
```

```python
#sort the skills
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
```
