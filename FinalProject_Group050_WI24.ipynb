## Dataset #1: Data Science Job Postings & Skills (2024)

```python
# importing all CSV files in the Job postings dataset
job_posting = pd.read_csv('job-postings/job_postings.csv')   
job_summaries = pd.read_csv('job-postings/job_summary.csv')   
job_skills = pd.read_csv('job-postings/job_skills.csv')   
```

```python
# Get relevant columns from the job postings csv file
job_desciptors = ['job_link','job_title', 'company', 'job_level', 'search_country']
df1 = job_posting[job_desciptors]

# Get relevant columns from the summaries csv file
df2 = job_summaries["job_summary"]

# Get relevant columns from the skills csv file
df3 = job_skills["job_skills"]

# Set 'job_link' as the index if it's not already, for the purpose of merging
df1.set_index('job_link', inplace=True)
df2 = job_summaries.set_index('job_link')
df3 = job_skills.set_index('job_link')

# Merge the three datasets together
jobs_data = df1.join([df2, df3])
jobs_data.reset_index(inplace=True)

# only keep the data for the United States
jobs_data = jobs_data[jobs_data['search_country'] == 'United States']

# look for null rows and drop them
null_rows = jobs_data[jobs_data.isnull().any(axis=1)] 
jobs_data.dropna()

# the jon_link column was only required for merging. Drop after merging is done 
jobs_data = jobs_data.drop('job_link', axis=1) 

# change column names
jobs_data.columns = ['job_title', 'company', 'job_level', 'country', 'summary', 'skill']

# CLEANING THE 'SKILL' COLUMN
jobs_data['skill'] = jobs_data['skill'].str.split(', ')

# Define the cleaning function to apply to each skill
def clean_skill_name(skill):
    skill_cleaned = re.sub(r'[^a-zA-Z0-9\s]', '', skill)  # Remove special characters
    skill_cleaned = skill_cleaned.lower()                # Convert to lowercase
    skill_cleaned = skill_cleaned.strip()                # Remove trailing whitespace
    return skill_cleaned
    
# Apply the cleaning function to each skill in each list in the 'skill' column
jobs_data['skill'] = jobs_data['skill'].apply(
    lambda skills: [clean_skill_name(skill)                  # for every skill in list , (not the entire list)
                    for skill in skills] 
                        if isinstance(skills, list)          # Handling the case for an empty list
                            else [])

# CLEANING THE "SUMMARY COLUMN"
def standardize_summary(summary):
    # Convert to lowercase
    summary = summary.lower()
    # Replace new lines and tabs with a space
    summary = re.sub(r'\n', ' ', summary)
    summary = re.sub(r'\t', ' ', summary)
    # Remove special characters (keeping only letters, numbers, and basic punctuation)
    summary = re.sub(r'[^a-zA-Z0-9 .,!?;]', '', summary)
    return summary
# Applying the standardization
jobs_data['summary'] = jobs_data['summary'].apply(standardize_summary)

jobs_data.to_csv('jobs_data.csv', index=False) 
jobs_data
```

## Dataset #2: Coursera Dataset

```python
# importing the Coursera dataset
courses = pd.read_csv("coursera-courses/CourseraDataset-Clean.csv")
courses.columns
```

```python
# getting the names of the columns that are needed
course_descriptors = ['Course Title', 'Rating', 'Level', 'Skill gain', 'Keyword', 'Number of Review']
courses_data = courses[course_descriptors]

# look for null rows (if any)
null_rows = courses_data[courses_data.isnull().any(axis=1)]    
print("Number of null rows: ", len(null_rows))    # there are no null rows

# Removing rows with non-ASCII characters (there were some rows with chinese letters)
mask = pd.Series([False] * len(courses_data))

# Check each column for non-ASCII characters
for column in courses_data.columns:
    # Update the mask to True if any column contains non-ASCII characters
    # This regex matches any character not in the ASCII range
    mask |= courses_data[column].astype(str).str.contains(r'[^\x00-\x7F]+', na=False)

# Dropping rows where any column contains non-ASCII characters 
courses_data = courses_data[~mask]
courses_data.reset_index(drop=True, inplace=True)

# Renaming the column names
courses_data.columns = ['course_title', 'rating', 'level', 'skill', 'keyword', 'num_review']

# getting all level values 
print("Different levels of courses in this dataset: ", courses_data['level'].unique())

# CLEANING THE "LEVEL" COLUMN 
def standardize_level(level):
    # Convert the level to lowercase and strip leading/trailing whitespace
    level = level.lower().strip()
    # Map each original level to its new standardized form
    if 'beginner level' in level:
        output = 'beginner'
    elif 'intermediate level' in level:
        output = 'intermediate'
    elif 'not specified' in level:
        output = 'N/A'
    elif 'advanced level' in level:
        output = 'advanced'
    # Otherwise, if uncaught - keep as is
    else:
        output = level
    
    return output
    
# Apply the function to the 'level' column
courses_data['level'] = courses_data['level'].apply(standardize_level)


# CLEANING THE "RATING" AND "NUM_REVIEW" COLUMN 
# ensure all numbers are floats or ints 
courses_data['rating'] = pd.to_numeric(courses_data['rating'], errors='coerce')
courses_data['num_review'] = pd.to_numeric(courses_data['num_review'], errors='coerce')

# CLEANING THE "KEYWORDS" COLUMN
unique_keywords = courses_data['keyword'].unique()
print("Keywords in this dataset are: ", unique_keywords)
# removing all courses that are NOT tech related 
exclude_keywords = ['Arts and Humanities', 'Business', 'Health', 'Personal Development',
                    'Physical Science and Engineering', 'Social Sciences']

# Corrected to 'keyword' to match the renamed column name
courses_data = courses_data[~courses_data['keyword'].isin(exclude_keywords)]

def standardize_keyword(keyw):
    # Convert the level to lowercase and strip leading/trailing whitespace
    keyw = keyw.lower().strip()
    # Map each original keyword to its new standardized form
    if 'computer science' in keyw:
        output = 'CS'
    elif 'datascience' in keyw:
        output = 'DS'
    elif 'information technology' in keyw:
        output = 'IT'
    elif 'math and logic' in keyw:
        output = 'MATH'
    # Otherwise, if uncaught - keep as is
    else:
        output = "N/A"
    
    return output
    
# Apply the function to the 'level' column
courses_data['keyword'] = courses_data['keyword'].apply(standardize_keyword)

# sorting the dataset first by num_reviews, then by rating
courses_data = courses_data.sort_values(["num_review", "rating"], ascending=False).reset_index(drop=True)


# CLEANING THE 'SKILL' COLUMN
def clean_skill_name(skill):
    """
    Clean the skill name based on specific rules.
    """
    replacements = {
        "python programming": "python",
        "r programming": "r",
        "cascading style sheets css": "css"
    }
    return replacements.get(skill.lower(), skill.lower())

# Cleaning the 'skill' column
courses_data['skill'] = courses_data['skill'].str.split(', ')
courses_data['skill'] = courses_data['skill'].apply(
    lambda skills: [clean_skill_name(skill) 
                    for skill in skills] 
                        if isinstance(skills, list) 
                            else []
)

courses_data
```

We will now define our function `calculate_skill_proportions` which takes in a dataframe of courses from the same sub category. This function aims to analyze how the teaching of the top 5 job skills (`python`, `sql`, `data analysis`, `machine learning`, `communication`) is distributed across different course subcategories within our Coursera dataset. In this function, we will filter the `sub_categories` DataFrame to only include courses that teach the top 5 skills. We count the numbers of such courses and store the value in `skill_counts`. Next, the total number of unique courses is stored in the `total_courses` variable. Finally, we calculate the ratio of _courses teaching each skill_.

This function is applied to each of the `subcategory_courses` dataframes - namely, `CS_courses`, `DS_courses`, `IT_courses`, and `MATH_courses` dataframes that we created in the previous cell.  

```python
def calculate_skill_proportions(subcategory_courses):
    skill_counts = subcategory_courses[subcategory_courses['skill'].isin(top_5_js)]['skill'].value_counts()
    total_courses = coursera_df['course_title'].nunique()
    skill_proportions = skill_counts / total_courses
    return skill_proportions

# applying the function to all subcategories 
CS_skill_proportions = calculate_skill_proportions(CS_courses)
DS_skill_proportions = calculate_skill_proportions(DS_courses)
IT_skill_proportions = calculate_skill_proportions(IT_courses)
MATH_skill_proportions = calculate_skill_proportions(MATH_courses)
MATH_skill_proportions
```

### Section 3.4: Popularity of Courses that Teach Highly Demanded Skills

In this section, we will look at the relationship between the number of reviews a course gets, and whether or not it is teaching a highly demanded data science job skill. To do so, we will first create a subset of our dataset that contains only those courses that teach a top-5-skill. We will use `.isin()` to filter the DataFrame by `top_5_js`. We will store this in `courses_with_top_skills`.

```python
courses_with_top_skills = coursera_df[coursera_df['skill'].isin(top_5_js)]
courses_with_top_skills
```

Next, we will count the number of courses that teach each of the top skills. We will use `.groupby()`, to accumulate the number of unique courses (using `.nunique()`) that teach each of the individual skills. We re-indexed the skills so they appear in the same order as before.

```python
# Count the number of courses teaching each of the top 5 job skills
courses_per_skill = coursera_df[coursera_df['skill'].isin(top_5_js)].groupby('skill')['course_title'].nunique()
courses_per_skill = courses_per_skill.reindex(top_5_js, fill_value=0)  # Ensure all top 5 skills are included, even if they have 0 courses
courses_per_skill
```

Next, we will group courses in the DataFrame `courses_with_top_skills` by the `skill` column. Furthermore, for every unique skill,  we will sum up the values in the `num_review` column.  This will result is a Series where each index is a unique skill and each value is the total number of reviews for courses related to that skill.

```python
# Counting the number of reviews per skills for the top 5 skills
reviews_per_skill = courses_with_top_skills.groupby('skill')['num_review'].sum()
reviews_per_skill = reviews_per_skill.reindex(top_5_js, fill_value=0)
reviews_per_skill
```

Here we can see that the total number of reviews on courses teaching python is the greatest. However, we cannot directly use this data for inference since the number of courses teaching each skill is different. Therefore, we need to see courses teaching which exact skill have proportionally higher number of reviews. To achieve this, we will first average the number of reviews by dividing `reviews_per_skill`  by `courses_per_skill`. We will store the result in `average_reviews_per_course`. 

Furthermore, we will combine `reviews_per_skill`, `courses_per_skill` and `average_reviews_per_course` into one DataFrame.

```python
# Calculate the average number of reviews per course for each skill
average_reviews_per_course = reviews_per_skill / courses_per_skill    
# combining into one dataframe for visualization
combined_data = pd.DataFrame({
    'Total Reviews': reviews_per_skill,
    'Number of Courses': courses_per_skill,
    'Average Reviews per Course': average_reviews_per_course
})
combined_data
```

Now, we must normalize the data before we visualize it so that we are able to compare all of them on the same scale. We will rescale the values in the dataset so that values are bound between 0 and 1, allowing for easier comparison between variables that may originally have different scales.

We shift every value in the DataFrame by doing `(combined_data - combined_data.min())` This shifts the values so that the lowest value is 0. Then, we divide by the range, which is given by `(combined_data.max() - combined_data.min())`  This step essentially does the re scaling. After this, the smallest value in each column will be 0, and the biggest value will be 1. 

The result, `combined_data_normalized`, is a new DataFrame where each value indicates the relative position of the original value within its column's range, rescaled to be between 0 and 1.

```python
# Normalize the data for visualization to compare the metrics on the same scale
combined_data_normalized = (combined_data - combined_data.min()) / (combined_data.max() - combined_data.min())
combined_data_normalized

palette = plt.cm.Pastel2.colors[:3]
fig, ax = plt.subplots(figsize=(7, 4))
combined_data_normalized.plot(kind='bar', ax=ax, width=0.7, color=palette )

# Add title and labels to the plot
ax.set_title('Comparison of Total Reviews, Number of Courses, and Average Reviews per Course for Top 5 Job Skills', fontsize=14)
ax.set_xlabel('Skill', fontsize=12)
ax.set_ylabel('Normalized Score', fontsize=12)
ax.legend(loc='upper right', fontsize=10)
plt.xticks(rotation=360)
plt.tight_layout()
plt.show()
```

From this graph, we can learn about the popularity of courses teaching each of the highly demanded skills:

- `python`: Has the highest number of total reviews and courses, Its reviews per skill is also high indicating a very high demand for python courses.
- `sql`: Has lower total reviews and courses than Python, and also a lower average reviews per course.
- `data analysis`: Has moderate total reviews and a higher average reviews per course, indicating that while there may be fewer courses, they are well-reviewed.

## Section 4: Significance Testing --  Are at least half of Coursera's Tech courses Industry-relevant?
Do 50% of the tech related courses that coursera offer teach at least the top 1 of the top 5 skills demanded by the jobs dataset?

- **Null hypothesis:** Proportion of coursera courses teaching top skills is less than or equal to 50%
- **Alternative hypothesis:** Proportion of coursera courses teaching top skills is greater than 50%

We begin with getting the number of tech-related Coursera courses that teach one of the top five skills. We also get the total number of tech-related Coursera courses. Using these two values, we get the proportion of Coursera tech courses that are "industry relevant".

```python
#get total number of tech-related cousera courses
total_courses_num = coursera_df['course_title'].nunique()
#get the courses that teach at least 1 of the top 5 skills
courses_that_teach_top_5 = courses_data[courses_data["skill"]
                                        .apply(lambda x: any(skill in x for skill in top_5_js))]
print("The total number of tech-related courses that Coursera offers is " + str(total_courses_num))
print("The total number of tech-related courses that Coursera offers, which also teach a top 5 skills is " 
                      + str(courses_that_teach_top_5.shape[0]))
print(str(int((courses_that_teach_top_5.shape[0]/total_courses_num)*100)) + "% of Coursera's Tech courses teach one of the top five skills!")
```

41% is a little bit lower than 50%. However, before we draw an conclusions we have to run our signficance test below. We used an alpha value of 0.05 as well as a one-sided z-test.

```python
alpha = 0.05
p_hat = courses_that_teach_top_5.shape[0] / total_courses_num        # Calculate the proportion of courses teaching top skills
p_null = 0.6                                                         # Perform one-sample proportion z-test
standard_error = np.sqrt(p_null * (1 - p_null) / total_courses)      # Calculate standard error
z_score = (p_hat - p_null) / standard_error                          # Calculate z-score

# Calculate p-value
p_value = 1 - stats.norm.cdf(z_score)                                # One-tailed test (greater than)

# Compare p-value with alpha
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

print("Proportion of courses teaching top skills:", p_hat)
print("Z-score:", z_score)
print("P-value:", p_value)
```

To reiterate, the null hypothesis stated that less than or equal to 50% of tech-related Coursera courses taught at least one of the top five job skills. The alternative hypothesis states that over 50% of tech-related Coursera courses taught at least one of the top five job skills.

This significance test aimed to see whether or not over 50% of tech-related cousera courses taught 1 of the top 5 skills that data science jobs required. Having performed a one-tailed test we failed to reject our null hypothesis that the percentages of tech-related courses that teach at least one top 5 skill is less than or equal to 50%.

Given a significance level (alpha) of 0.05, we compare the p-value to this threshold. Since the p-value of 1.0 is greater than the alpha value, we fail to reject the null hypothesis.

However, it is important to consider that not all tech-related courses are beginner level courses. Some are targeted at more advanced programmers and thus would not include labels such as "python programming" or "data visualization" as those skills are most likely deemed as pre-requisites for higher complexity courses.

## Exploring further: Are Data Science Skills taught only in Data Science Classes?

From the Distribution shown above, we are able to visualize the proportion of the most demanded skills taught across different sub-categories of tech courses - Data Science, IT, Computer Science, and Math. We will now explore whether people should only stick to a "Data Science" curriculum to land a Data Science Job, based on the 2024 Job postings from Linkedin. This will further give us insights into the **significance of having an interdisciplinary education**, for working in the industry. 

We will explore, to what extent are the most demanded skills for data science positions (on Linkedin), represented within the curricula of courses labeled as "Data Science" compared to their representations across a broader spectrum of _technology-related_ courses on Coursera. 

To achieve this, we will conduct a **Chi Squared Contingency Test**. It will help us **determine whether there is a significant association between the type of course (Data Science vs. other tech-related courses) and the inclusion of the top skills**.

We will begin by accumulating the counts of the skills taught in each of the categories. For example, how many CS courses teach python, how many DS courses teach python and so on.

```python
# Counting skills taught in each sub-category
CS_skill_counts = CS_courses['skill'].value_counts()
DS_skill_counts = DS_courses['skill'].value_counts()
IT_skill_counts = IT_courses['skill'].value_counts()
MATH_skill_counts = MATH_courses['skill'].value_counts()

# a new data frame that combines these counts into a table
course_skills_table = pd.DataFrame({
    'CS': CS_skill_counts,
    'DS': DS_skill_counts,
    'IT': IT_skill_counts,
    'MATH': MATH_skill_counts
}).reindex(top_5_js).fillna(0)  # use reindex to ensure the skills are in the order of the list

course_skills_table
```

We will now conduct a `chi2_congtingency` test. 

Let's break this down:
1. `chi2` is the Chi-square statistic, a measure that tells how expectations compare to actual observed data. 
2. `p` stores the p-value, which tells you the probability of observing the data assuming that the null hypothesis is true.
3. `dof` is the degrees of freedom. We have 12 degrees of freedom because we have 5 skills and 4 keywords. So, dof = (5-1)(4-1) = 12/
4. `expected` contains the expected frequencies of each cell in the table if the null hypothesis were true.

```python
chi2, p, dof, expected = chi2_contingency(course_skills_table)

print("Chi-Square Statistic is: ", chi2, "\n")
print("The p-value is: ", p, "\n")
print("The Degrees of freedom are: ", dof, "\n")
print("Expected frequencies:\n ", expected, "\n")

if p < 0.01:
    print('There is a significant difference in the distribution of top skills across course categories!')
else:
    print('There is no significant difference in the distribution of top skills across course categories.')
```

Since the p-value is less than 0.01, we can reject the null hypothesis. This implies that there **is a statistically significant association between the type of course (Data Science, Computer Science, IT, Math) and the inclusion of the top skills**. That is, the distribution of skills across the different course categories is not random, and certain skills are more likely to be taught in certain categories. 

For example, the skill "data analysis" appears to be more prevalent in DS courses than in other categories, which could be due to the nature of data science, which heavily relies on analyzing data. This significant difference indicates that **Coursera's course offerings are tailored to the specific skill demands** of each tech field. People interested in acquiring a particular skill -- maybe to land a Data Science Job --  would enroll in courses from a specific department. Such insights help us in evaluating the **alignment of course content with market demands**.
